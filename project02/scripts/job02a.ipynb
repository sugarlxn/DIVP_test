**cmap**: str or Colormap, default: rcParams["image.cmap"] (default: 'viridis')
The Colormap instance or registered colormap name used to map scalar data to colors.

This parameter is ignored if X is RGB(A).

````python
#!/usr/bin/python3 
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
````

````python
root_path = '/root/DIVP_project/'
einstein_path = "project02/proj02-images/einstein_orig.tif"
baboon_path = "project02/proj02-images/baboon.jpg"
````

````python
image_einstein = Image.open(root_path + einstein_path)
plt.imshow(image_einstein,cmap='gray')
````

````python
# 以不同的cmap 选项显示图片
plt.figure()
plt.subplot(331)
plt.imshow(image_einstein,cmap='gray')
plt.title('Original')
plt.subplot(332)
plt.imshow(image_einstein,cmap='hot')
plt.title('hot')
plt.subplot(333)
plt.imshow(image_einstein,cmap='spring')
plt.title('spring')
plt.subplot(334)
plt.imshow(image_einstein,cmap='cool')
plt.title('cool')
plt.subplot(335)
plt.imshow(image_einstein,cmap='winter')
plt.title('winter')
plt.subplot(336)
plt.imshow(image_einstein,cmap='autumn')
plt.title('autumn')
plt.subplot(337)
plt.imshow(image_einstein,cmap='summer')
plt.title('summer')
plt.subplot(338)
plt.imshow(image_einstein,cmap='Greens')
plt.title('Greens')
plt.subplot(339)
plt.imshow(image_einstein,cmap='Blues')
plt.title('Blues')
plt.tight_layout()
plt.show()
````
