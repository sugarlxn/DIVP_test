读取一幅灰度图像，实现图像灰度的对数变换，在屏幕上显示变换前、后图像，以及变换前后图像的
灰度直方图。

```python
#!/usr/bin/python3 
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
root_path = '/root/DIVP_project/'
einstein_path = "project02/proj02-images/einstein_orig.tif"
```

```python
# 读取一张灰度图片
image_einstein = Image.open(root_path + einstein_path)
print(type(image_einstein))
print(image_einstein.size)
print(image_einstein.mode)
```

```python
# 编写一个函数 image2log, 将图像的像素值转换为对数值
def image2log(image):
    image = np.array(image)
    image = np.log(image + 1)
    image = np.uint8(image)
    image = Image.fromarray(image)
    return image
```

```python
# 对image_einstein 进行对数变换
image_einstein_log = np.array(image_einstein)
image_einstein_log = np.log(image_einstein_log + 1)
image_einstein_log = np.uint8(image_einstein_log)
image_einstein_log = Image.fromarray(image_einstein_log)
```

```python
# 使用自己的方法image2log 对图像进行对数变换
image_einstein_log2 = image2log(image_einstein)
```

```python
# 统计灰度的图像的直方图
plt.figure()
plt.subplot(221)
plt.imshow(image_einstein, cmap='gray')
plt.title('original image')
plt.subplot(222)
plt.hist(np.array(image_einstein).flatten(), bins=256, range=[0, 256], color='r')
plt.xlabel('gray value')
plt.ylabel('number of pixels')
plt.title('histogram of gray value')
plt.subplot(223)
plt.imshow(image_einstein_log2, cmap='gray')
plt.title('log image')
plt.subplot(224)
plt.hist(np.array(image_einstein_log2).flatten(), bins=256, range=[0, 256], color='r')
plt.xlabel('gray value')
plt.ylabel('number of pixels')
plt.title('histogram of gray value')
plt.tight_layout()
plt.show()
```
