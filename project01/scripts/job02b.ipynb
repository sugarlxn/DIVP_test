读取一幅灰度图像，将其灰度级减少1/2，并显示。重复上述过程，直到该灰度图像变成二值图像。
比较这些图像，说明灰度级缩减对图像有什么影响。（提示，可以使用矩阵的整数乘、除法进行灰度缩减）

```python
#!/usr/bin/python3 
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
root_path = '/root/DIVP_project/'
image_path_vivid = "project01/proj01-images/lena_std.bmp"
image_path_gray = "project01/proj01-images/Fig2.22(b).jpg"
image_path_tif = "project01/proj01-images/Fig0308(a)(fractured_spine).tif"
image_path_moon = "project01/proj01-images/Fig0338(a)(blurry_moon).tif"
```

```python
image_gray = Image.open(root_path + image_path_moon)
print(type(image_gray))
print(image_gray.size)
print(image_gray.mode)
```

```python
# 将 image_gray 转换为 numpy 数组
image_gray = np.array(image_gray)
print(type(image_gray))
print(image_gray.shape)
print(image_gray.dtype)
```

```python
def down_scale_gray(image, scale):
    image_temp = image.copy()
    image_temp = image_temp // (2 ** scale)
    image_result = image_temp * (2 ** scale)
    image_result = Image.fromarray(image_result)
    return image_result
```

```python
image_temp = image_gray // 2
image_result = image_temp * 2
image_result = Image.fromarray(image_result)
plt.imshow(image_result, cmap='gray')
```

```python
image_1 = down_scale_gray(image_gray, 1)
image_2 = down_scale_gray(image_gray, 2)
image_3 = down_scale_gray(image_gray, 3)
image_4 = down_scale_gray(image_gray, 4)
image_5 = down_scale_gray(image_gray, 5)
image_6 = down_scale_gray(image_gray, 6)
image_7 = down_scale_gray(image_gray, 7)

figure = plt.figure()
plt.subplot(241)
plt.imshow(image_gray, cmap='gray')
plt.title('Original')
plt.subplot(242)
plt.imshow(image_1, cmap='gray')
plt.title('Down scale 1')
plt.subplot(243)
plt.imshow(image_2, cmap='gray')
plt.title('Down scale 2')
plt.subplot(244)
plt.imshow(image_3, cmap='gray')
plt.title('Down scale 3')
plt.subplot(245)
plt.imshow(image_4, cmap='gray')
plt.title('Down scale 4')
plt.subplot(246)
plt.imshow(image_5, cmap='gray')
plt.title('Down scale 5')
plt.subplot(247)
plt.imshow(image_6, cmap='gray')
plt.title('Down scale 6')
plt.subplot(248)
plt.imshow(image_7, cmap='gray')
plt.title('Down scale 7')
plt.tight_layout()
plt.show()
```
