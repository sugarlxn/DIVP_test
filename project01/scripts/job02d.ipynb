读取二幅合适的灰度图像，将其进行alpha混合加法运算，并显示。

```python
#!/usr/bin/python3 
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
root_path = '/root/DIVP_project/'
image_path_vivid = "project01/proj01-images/lena_std.bmp"
image_path_gray = "project01/proj01-images/Fig2.22(b).jpg"
image_path_tif = "project01/proj01-images/Fig0308(a)(fractured_spine).tif"
```

```python
image1 = cv2.imread(root_path + image_path_vivid, cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread(root_path + image_path_gray, cv2.IMREAD_GRAYSCALE)

# 将两张图片resize到相同大小
image1 = cv2.resize(image1, (image2.shape[1], image2.shape[0]))
```

```python
# 定义一个函数，对image1 和image2 队员元素求 加权和
def weighted_sum(image1, image2, alpha):
    # 定义一个和image1 一样大小的全0矩阵
    result = np.zeros(image1.shape)
    for i in range(image1.shape[0]):
        for j in range(image1.shape[1]):
            result[i][j] = alpha * image1[i][j] + (1 - alpha) * image2[i][j]
    return result
```

```python
# 对image1 与 image2 进行alpha混合加法运算
# alpha = 0.7
# beta = 0.3
# gamma = 0
# result = cv2.addWeighted(image1, alpha, image2, beta, gamma)
result = weighted_sum(image1, image2, 0.7)
```

```python
# 显示结果
plt.subplot(131)
plt.imshow(image1, cmap='gray')
plt.title('image1')
plt.subplot(132)
plt.imshow(image2, cmap='gray')
plt.title('image2')
plt.subplot(133)
plt.imshow(result, cmap='gray')
plt.title('result')
plt.tight_layout()
plt.show()
```
